# Min-Max Normalisation

```python
#MIN MAX NORMALIZATION
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
```

```python
#Create an object to transform the data to fit minmax processor
x_scaled = min_max_scaler.fit_transform(quality[['DaysSinceLastERVisit']].values.astype(float))
```

```python
#Run the normalizer on the dataframes
quality.normalized = pd.DataFrame(x_scaled)
quality['normalized2']=x_scaled
quality.sort_values(by='DaysSinceLastERVisit')
quality[['DaysSinceLastERVisit','normalized2']]
```

# Naive Bayes Classification

```python
quality.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
encoded=le.fit_transform(quality['StartedOnCombination'])
quality['encoded_Combination']=encoded
print(quality)
```

```python
quality.columns
```

```python
quality.drop(['StartedOnCombination','normalized2', 'zscore','MemberID'],axis='columns', inplace=True)
```

```python
quality.shape
```

```python
quality.columns
```

```python
Y=quality[['PoorCare']]
X=quality.drop(columns=['PoorCare'])
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
y_train.value_counts()
```

```python
x_train.value_counts()
```

```python
x_train.shape
```

```python
#SMOTE-Synthetic Minority Oversampling Technique
from imblearn.over_sampling import SMOTE
sm = SMOTE()
x_train_res, y_train_res = sm.fit_resample(x_train, y_train)
```

```python
x_train_res.shape
```

```python
y_train_res.shape
```

```python
y_train_res.value_counts()
```

```python
from sklearn.naive_bayes import GaussianNB
naive=GaussianNB()
```

```python
naive_model=naive.fit(x_train_res,y_train_res)
naive_model
print("The model has been executed successfully")
```

```python
y_test
```

```python
x_test.shape
```

```python
y_test['Predicted']=naive_model.predict(x_test)
```

```python
y_test
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(confusion_matrix(y_test['PoorCare'],y_test['Predicted']))
```

```python
accuracy=accuracy_score(y_test['PoorCare'],y_test['Predicted'])
accuracy=np.round(accuracy,2)
print("Accuracy of the model is : ",accuracy)
```

```python
print(classification_report(y_test['PoorCare'],y_test['Predicted']))
```

# Decision Tree

```python
#using decision tree model
from sklearn.tree import DecisionTreeClassifier
regressor=DecisionTreeClassifier(criterion='gini',max_depth=5,splitter='best')
X
regressor.fit(x_train_res,y_train_res)#training of classifier
y_pred=regressor.predict(val_data.transpose())
print("Prediction: ",y_pred)
```

```python
y_test['Predicted_dt']=regressor.predict(x_test)
```

```python
y_test
```

```python
print(confusion_matrix(y_test['PoorCare'],y_test['Predicted_dt']))
```

```python
accuracy=accuracy_score(y_test['PoorCare'],y_test['Predicted_dt'])
accuracy=np.round(accuracy,2)
print("Accuracy of the model is : ",accuracy)
```

```python
print(classification_report(y_test['PoorCare'],y_test['Predicted_dt']))
```
